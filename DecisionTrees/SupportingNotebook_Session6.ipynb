```python
# Classification Metrics

import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# True labels (actual values)
y_true = np.array([0, 0, 0, 1, 1, 0, 1, 1, 0, 1])

# Predicted labels (predictions from a classifier)
y_pred = np.array([0, 0, 0, 1, 1, 0, 1, 1, 1, 0])

# Accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f"Accuracy: {accuracy}")

# Precision
precision = precision_score(y_true, y_pred)
print(f"Precision: {precision}")

# Recall
recall = recall_score(y_true, y_pred)
print(f"Recall: {recall}")

# F1-Score
f1 = f1_score(y_true, y_pred)
print(f"F1-Score: {f1}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_true, y_pred)
labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / conf_matrix.flatten().sum())]
            for item in conf_matrix.flatten()
        ]
    ).reshape(2, 2)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=labels, fmt="")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.show()

# TP = 4
# TN = 4
# FP = 1
# FN = 1

# True labels (actual values)
y_true = np.array([0, 0, 0, 1, 1, 0, 1, 1, 0, 1])

# Predicted labels (predictions from a classifier)
y_pred = np.array([0, 0, 0, 1, 1, 0, 1, 1, 1, 0])
# Prec = tp / tp + fp
```

```python
# Use of cross tab

import pandas as pd
data = {
    'Category': ['A', 'B', 'C', 'A', 'A', 'B', 'B', 'C'],
    'Target': ['True', 'True', 'False', 'True', 'True', 'False', 'True', 'False']
}
pd.crosstab(data['Category'], data['Target'])
```

```python
pd.crosstab(data['Category'], data['Target'], normalize="index") # %age across rows
```

```python
# One Hot Encoding

# Create a DataFrame with a categorical column 'Color'
data = {
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Blue', 'Red'],
    'Size': ['M', 'L', 'S', 'M', 'L', 'S'],
    'Price': [10, 20, 30, 25, 35, 40]
}

df = pd.DataFrame(data)

df_dummies = pd.get_dummies(df) # , columns=['Color', 'Size']

print(df_dummies)
```

```python
# Get dummies for the 'Color' column
df_dummies = pd.get_dummies(df, drop_first=True)

print(df_dummies)
```
